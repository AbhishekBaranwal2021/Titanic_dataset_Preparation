```python
train['Fare'] = train['Fare'].fillna(train['Fare'].mean()) # Changing the null values to mean valus
train['Age'] = train['Age'].fillna(train['Age'].mean()) 
train['Embarked'] = train['Embarked'].fillna(method='ffill') #changing the null values to forward fill
```

```python
train.isnull().sum()
```

```python
train.drop(['Ticket','Cabin'], axis=1,inplace=True) # droping the values which have contain high no of null values
```
